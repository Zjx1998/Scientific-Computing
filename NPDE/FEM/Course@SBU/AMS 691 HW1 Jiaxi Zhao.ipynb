## Lagrange basis functions for master element
### 2D master element
Implement a subroutine to construct the shape functions over a triangular master
element with corner nodes $\xi_1 = (0, 0), \ \xi_2 = (1, 0), \ \xi_3 = (0, 1)$ for an arbitrary $p$. Plot th generated Lagrange basis functions for $p = 1, \cdots, 4$.

To illustrate the result, we also print their formula abvoe the figure.

```python
# Construction of shape function for triangular master element
x, y = sp.symbols("x y")

for p in range(2, 5):
    node_num = 0
    num      = int((p + 1) * (p + 2) / 2)  # number of the basis
    node     = np.zeros([num, 2])          # node matrix stores the coordinate of nodes 
    V        = np.zeros([num, num])        # Vandemonde matrix
    basis_f  = []                          # basis function
    
    # construct the node matrix and basis function
    for i in range(p + 1):
        for j in range(p - i + 1):
            node[node_num, 0]  = i / p
            node[node_num, 1]  = j / p
            basis_f.append(pow(x, i) * pow(y, j))
            node_num           = node_num + 1
            
    # Construct the Vandemonde matrix
    for i in range(node_num):
        for j in range(node_num):
            V[i, j] = basis_f[j].evalf(subs ={'x' : node[i, 0], 
                                              'y' : node[i, 1]})
    inv_V = np.linalg.inv(V)

    print("Lagrangian basis in 2D triangular master element of order", p)
    for i in range(1, node_num):
        func  = inv_V[i, :].dot(basis_f)
        print(func)
        plot3d_parametric_surface(x, y, func, (x, 0, 1), (y, 0, 1))
    print("")
    
```

### 3D master element
Implement a subroutine to construct the shape functions over a triangular master
element with corner nodes $\xi_1 = (0, 0, 0), \ \xi_2 = (1, 0, 0), \ \xi_3 = (0, 1, 0), \xi_4 = (0, 0, 1)$ for an arbitrary $p$. To illustrate this result, we only state the formula for each basis function.

```python
# Construction of shape function for triangular master element
x, y, z = sp.symbols("x y z")

for p in range(1, 4):
    node_num = 0
    num      = int((p + 1) * (p + 2) * (p + 3) / 6)  # number of the basis
    node     = np.zeros([num, 3])          # node matrix stores the coordinate of nodes 
    V        = np.zeros([num, num])        # Vandemonde matrix
    inv_V    = np.zeros([num, num])        # inverse Vandemonde matrix
    basis_f  = []                          # basis function
    
    # construct the node matrix and basis function
    for i in range(p + 1):
        for j in range(p - i + 1):
            for k in range(p - i - j + 1):
                node[node_num, 0]  = i / p
                node[node_num, 1]  = j / p
                node[node_num, 2]  = k / p
                basis_f.append(pow(x, i) * pow(y, j) * pow(z, k))
                node_num           = node_num + 1

    # Construct the Vandemonde matrix
    for i in range(node_num):
        for j in range(node_num):
            V[i, j] = basis_f[j].evalf(subs ={'x' : node[i, 0], 
                                              'y' : node[i, 1], 
                                              'z' : node[i, 2]})
    inv_V = np.linalg.inv(V)

    print("Lagrangian basis in 3D triangular master element of order", p)
    for i in range(node_num):
        print(inv_V[i,:].dot(basis_f))
    print("")
```

After constructing the Lagrangian basis for the master element in 3D, we show here that for order $p = 3$, this functions can approximate any polynomial with degree 3.

To verify this, we randomly generated a function which is given by $f = c^T F$ with $F$ the set of Lagrangian basis function, i.e. $f(x) = \sum_{i = 1}^n c_i F_i(x)$. Here $n$ equals to the dimension of the polynomial spaces with degree less than 4 on three variables $x, y, z$. Evaluate this function at nodes provides a vector $v \in \mathbb{R}^n$, which is ecatly
\begin{equation}
    v_j = \sum_{i = 1}^n c_i F_i(x_j) = \sum_{i = 1}^n c_i V_{ij} = (c^T V)_j.
\end{equation}
The $V$ above is the Vandemonde matrix. And since we obtain the interpolation function by letting the parameter of each basis function be $\widetilde{c}^T = V^{-1}v = c^T$, hence interpolate exactly.
